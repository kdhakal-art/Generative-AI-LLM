```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\kdhakal\Documents\NLP\tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove("not")
```

apply() : takes one column and perform a custom operation on every value in it.

```python
data['review_no_stopwords'] = data['review_lowercase'].apply(lambda x: ' '.join(word for word in x.split() if word not in en_stopwords))
```

```python
data['review_no_stopwords'][0]
```

```python
data['review_no_stopwords_no_punct'] = data.apply(lambda x: re.sub(r'[*]','star',x['review_no_stopwords']), axis=1)
```

```python
data['review_no_stopwords_no_punct'][0]
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punct'] = data.apply(lambda x: re.sub(r'[^\w\s]','',x['review_no_stopwords_no_punct']), axis=1)
```

```python
data.head()
```

Text tokenization

```python
data['tokenized'] = data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punct']), axis=1)
```

```python
data['tokenized'][0]
```

```python
ps = PorterStemmer()
```

```python
data['stemmed'] = data['tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
data['lemmatized'] = data['tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
data['lemmatized'][0]
```

Each row in the 'lemmatized' column contains a list of lemmatized words from the corresponding review text after preprocessing.
Each review is stored as a list of tokens in the 'lemmatized' column.
We need to combine these smaller lists into one long list that contains every token from all reviews.

```python
tokens_clean = sum(data['lemmatized'], [])
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
```

```python
print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
print(bigrams)
```
