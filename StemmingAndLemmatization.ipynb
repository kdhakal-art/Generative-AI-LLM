```python
#Stemming reduces the number of unique words
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
connect_tokens = ['connect', 'connecting', 'connected', 'connects', 'connectivity']
```

```python
for t in connect_tokens:
    print(t, ": ", ps.stem(t))
```

Lemmatization

```python
#Lemmatization reduces a word to a meaningful base form while preserving its meaning
```

```python
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
for t in connect_tokens:
    print(t, ": ", WordNetLemmatizer().lemmatize(t))
```

N-gram : help us analyze the relationship between neighboring words
N-gram = a sequence of N tokens
N=1 : unigrams
N=2 : bigram
N=n : N-gram

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tokens = ['the', 'rain', 'in', 'spain', 'falls', 'mainly', 'on', 'the', 'plain', 'the', 'rise', 'of', 'tokenization', 'in', 'ai']
print(tokens)
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens, 1)).value_counts())
```

```python
print(unigrams)
```

```python
unigrams[:10].sort_values().plot.barh(color='lightsalmon')
plt.title('most frequent unigrams')
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())
print(bigrams)
```
