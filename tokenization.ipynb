```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentences = "Her cat name is Luna. Her dog name is Max."
```

```python
sent_tokenize(sentences)
```

```python
sentence = "Her cat name is Luna"
word_tokenize(sentence)
```

```python
sentence_1 = "her cat name is Luna. Her dog name is Max."
word_tokenize(sentence_1.lower())
```
